```python
# Import Required Libraries
import os
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import jams

print("✅ All libraries imported successfully")

# Define Paths
FEATURES_PATH = "C:/Users/chatt/Desktop/GuiterDataset-20251203T044716Z-3-001/GuiterDataset/features"
ANNOTATIONS_PATH = "C:/Users/chatt/Desktop/GuiterDataset-20251203T044716Z-3-001/GuiterDataset/annotation"
print("📁 Paths defined")

X, y = [], []

print("🚀 Loading and preprocessing data...")
# Load and Preprocess Data
for file in os.listdir(FEATURES_PATH):
    if not file.endswith(".npz"): continue
    base_name = file[:-4]
    base_name = base_name[:-4] if base_name.endswith("_mic") else base_name

    feat_file = os.path.join(FEATURES_PATH, file)
    annotation_file = os.path.join(ANNOTATIONS_PATH, base_name + ".jams")
    if not os.path.exists(annotation_file):
        print(f"⚠️ Missing annotation for {file}")
        continue

    data = np.load(feat_file)
    feature_stack = np.concatenate([data['hpcp'], data['chroma'], data['tonnetz']], axis=1)

    jam = jams.load(annotation_file)
    chords_ann = jam.search(namespace="chord")[0]
    intervals = np.array([[ann.time, ann.time + ann.duration] for ann in chords_ann.data])
    chord_labels = [ann.value for ann in chords_ann.data]

    frame_times = np.arange(feature_stack.shape[0]) * (512 / 22050.0)
    frame_chords = []
    for t in frame_times:
        idx = np.where((t >= intervals[:, 0]) & (t <= intervals[:, 1]))[0]
        frame_chords.append(chord_labels[idx[0]] if len(idx) > 0 else "N")

    X.append(feature_stack)
    y.append(frame_chords)

print(f"✅ Processed {len(X)} files")

# Convert to single arrays
print("🔄 Converting lists to arrays...")
X = np.vstack(X)
y = np.concatenate(y)

print("🔠 Encoding labels...")
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print("✂️ Splitting data into training and test sets...")
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

print("📐 Reshaping data for LSTM input...")

# Step 1: Calculate how many full chunks we can get
frame_len = 500
feature_dim = X.shape[1]

usable_train = (X_train.shape[0] // frame_len) * frame_len
usable_test = (X_test.shape[0] // frame_len) * frame_len

X_train = X_train[:usable_train].reshape(-1, frame_len, feature_dim)
X_test = X_test[:usable_test].reshape(-1, frame_len, feature_dim)
y_train = y_train[:usable_train].reshape(-1, frame_len)
y_test = y_test[:usable_test].reshape(-1, frame_len)

print(f"✅ Reshaped X_train to {X_train.shape}, y_train to {y_train.shape}")


print("🧠 Building the model...")
model = tf.keras.Sequential([
    tf.keras.layers.Conv1D(192, kernel_size=3, activation='relu', padding='same', input_shape=(500, X.shape[1])),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128, return_sequences=True)),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(len(le.classes_), activation='softmax'))
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary(print_fn=lambda x: print("📝 " + x))

print("🏋️‍♂️ Training started...")
history = model.fit(
    X_train, y_train, 
    validation_data=(X_test, y_test), 
    epochs=20,
    batch_size=32
)
print("✅ Training completed")

print("🔍 Evaluating model...")
pred = model.predict(X_test)
pred_classes = np.argmax(pred, axis=-1)
mask = y_test.flatten() != -1

print("\n📊 Classification Report:")
print(classification_report(y_test.flatten()[mask], pred_classes.flatten()[mask], target_names=le.classes_))

print("📈 Plotting Accuracy Graph...")
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

print("📉 Plotting Confusion Matrix...")
cm = confusion_matrix(y_test.flatten()[mask], pred_classes.flatten()[mask])
plt.figure(figsize=(12, 10))
sns.heatmap(cm, xticklabels=le.classes_, yticklabels=le.classes_, cmap='Blues', fmt='d')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
# --- Ensemble Chord Recognition Model: CNN-BiLSTM + RandomForest ---

import os
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
import tensorflow as tf
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
import jams

print("✅ All libraries imported successfully")

# --- Define Paths ---
FEATURES_PATH = "C:/Users/chatt/Desktop/GuiterDataset-20251203T044716Z-3-001/GuiterDataset/features"
ANNOTATIONS_PATH = "C:/Users/chatt/Desktop/GuiterDataset-20251203T044716Z-3-001/GuiterDataset/annotation"
print("📁 Paths defined")

X_seq, y_seq, X_flat, y_flat = [], [], [], []
print("🚀 Loading and preprocessing data...")

# --- Load and preprocess data ---
for file in os.listdir(FEATURES_PATH):
    if not file.endswith(".npz"): continue
    base_name = file[:-4]
    base_name = base_name[:-4] if base_name.endswith("_mic") else base_name

    feat_file = os.path.join(FEATURES_PATH, file)
    annotation_file = os.path.join(ANNOTATIONS_PATH, base_name + ".jams")
    if not os.path.exists(annotation_file):
        print(f"⚠️ Missing annotation for {file}")
        continue

    data = np.load(feat_file)
    features = np.concatenate([data['hpcp'], data['chroma'], data['tonnetz']], axis=1)

    jam = jams.load(annotation_file)
    chords_ann = jam.search(namespace="chord")[0]
    intervals = np.array([[ann.time, ann.time + ann.duration] for ann in chords_ann.data])
    chord_labels = [ann.value for ann in chords_ann.data]

    frame_times = np.arange(features.shape[0]) * (512 / 22050.0)
    frame_chords = []
    for t in frame_times:
        idx = np.where((t >= intervals[:, 0]) & (t <= intervals[:, 1]))[0]
        frame_chords.append(chord_labels[idx[0]] if len(idx) > 0 else "N")

    X_seq.append(features)
    y_seq.append(frame_chords)

    # Statistical summary for RF
    summary = np.hstack([features.mean(axis=0), features.std(axis=0)])
    X_flat.append(summary)
    y_flat.append(max(set(frame_chords), key=frame_chords.count))

print(f"✅ Processed {len(X_seq)} files")

# --- Convert and Encode ---
X_all = np.vstack(X_seq)
y_all = np.concatenate(y_seq)

le = LabelEncoder()
y_encoded = le.fit_transform(y_all)

# --- Prepare Sequential Data for Deep Model ---
frame_len = 500
feat_dim = X_all.shape[1]
usable = (X_all.shape[0] // frame_len) * frame_len

X_seq_final = X_all[:usable].reshape(-1, frame_len, feat_dim)
y_seq_final = y_encoded[:usable].reshape(-1, frame_len)

X_train_seq, X_test_seq, y_train_seq, y_test_seq = train_test_split(X_seq_final, y_seq_final, test_size=0.2, random_state=42)

# --- Prepare Flat Features for Random Forest ---
X_rf = np.array(X_flat)
y_rf = le.transform(y_flat)
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(X_rf, y_rf, test_size=0.2, random_state=42)

# --- Hyperparameter Tuning for Random Forest ---
print("🔍 Tuning Random Forest hyperparameters...")
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
}
rf_tuner = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3, scoring='accuracy', n_jobs=-1)
rf_tuner.fit(X_train_rf, y_train_rf)
rf = rf_tuner.best_estimator_
print(f"✅ Best RF Params: {rf_tuner.best_params_}")

# --- Hyperparameter Tuning for Deep Learning Model ---
print("🧠 Tuning CNN-BiLSTM model...")
best_dnn = None
best_val_acc = 0
for dropout in [0.3, 0.4]:
    for lstm_units in [64, 128]:
        model = tf.keras.Sequential([
            tf.keras.layers.Input(shape=(frame_len, feat_dim)),
            tf.keras.layers.Conv1D(192, kernel_size=3, activation='relu', padding='same'),
            tf.keras.layers.BatchNormalization(),
            tf.keras.layers.Dropout(dropout),
            tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(lstm_units, return_sequences=True)),
            tf.keras.layers.Dropout(dropout),
            tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(len(le.classes_), activation='softmax'))
        ])
        model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
        history = model.fit(X_train_seq, y_train_seq, validation_data=(X_test_seq, y_test_seq), epochs=10, batch_size=32, verbose=0)
        val_acc = history.history['val_accuracy'][-1]
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_dnn = model

print(f"✅ Best DNN Validation Accuracy: {best_val_acc:.4f}")

# --- Predictions ---
dnn_preds = best_dnn.predict(X_test_seq)
dnn_segment_probs = [segment.mean(axis=0) for segment in dnn_preds]

rf_probs = rf.predict_proba(X_test_rf)
rf_class_indices = rf.classes_
rf_probs_aligned = []
for probs in rf_probs:
    aligned = np.zeros(len(le.classes_))
    for rf_class_idx, prob in zip(rf_class_indices, probs):
        aligned[rf_class_idx] = prob
    rf_probs_aligned.append(aligned)
rf_probs_aligned = np.array(rf_probs_aligned)

# --- Blending ---
final_preds = []
alpha = 0.6
for dnn_prob, rf_prob in zip(dnn_segment_probs, rf_probs_aligned):
    blended = alpha * dnn_prob + (1 - alpha) * rf_prob
    final_preds.append(np.argmax(blended))

# --- Evaluation ---
rf_preds = rf.predict(X_test_rf)
rf_accuracy = (rf_preds == y_test_rf).mean()
dnn_file_preds = np.argmax(dnn_segment_probs, axis=1)
# Use the most common chord per segment in y_test_seq for fair comparison
y_test_seq_major = [np.bincount(segment).argmax() for segment in y_test_seq]
dnn_accuracy = (dnn_file_preds == y_test_seq_major).mean()

ensemble_accuracy = (np.array(final_preds) == y_test_rf).mean()

print("\n📊 Classification Report:")
labels_used = sorted(np.unique(np.concatenate([y_test_rf, final_preds])))
target_names_used = le.inverse_transform(labels_used)
print(classification_report(y_test_rf, final_preds, labels=labels_used, target_names=target_names_used))


# --- Accuracy Plot ---
accs = [rf_accuracy, dnn_accuracy, ensemble_accuracy]
plt.bar(['Random Forest', 'CNN-BiLSTM', 'Ensemble'], accs)
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.show()

print(f"\n📈 Individual Model Accuracies:")
print(f"Random Forest: {rf_accuracy:.4f}")
print(f"CNN-BiLSTM: {dnn_accuracy:.4f}")
print(f"Ensemble: {ensemble_accuracy:.4f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, ConfusionMatrixDisplay, r2_score
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

# --- Encode labels ---
le = LabelEncoder()
y_encoded = le.fit_transform(y_flat)

# --- Normalize features ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(np.array(X_flat))

# --- Reshape for CNN ---
X_cnn = X_scaled.reshape(-1, X_scaled.shape[1], 1)

# --- Split data ---
X_train, X_test, y_train, y_test = train_test_split(
    X_cnn, y_encoded, test_size=0.2, random_state=42
)

# --- RF Hyperparameter Tuning ---
X_rf_train, X_rf_test = X_train.reshape(X_train.shape[0], -1), X_test.reshape(X_test.shape[0], -1)
rf_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'bootstrap': [True],
    'max_features': ['sqrt']
}
rf_cv = GridSearchCV(RandomForestClassifier(random_state=42, class_weight='balanced'), rf_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
rf_cv.fit(X_rf_train, y_train)
rf_best = rf_cv.best_estimator_
rf_train_preds = rf_best.predict(X_rf_train)
rf_test_preds = rf_best.predict(X_rf_test)

# --- CNN Model ---
cnn_model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train.shape[1], 1)),
    tf.keras.layers.Conv1D(64, kernel_size=3, activation='relu'),
    tf.keras.layers.MaxPooling1D(pool_size=2),
    tf.keras.layers.Conv1D(128, kernel_size=3, activation='relu'),
    tf.keras.layers.GlobalMaxPooling1D(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(len(le.classes_), activation='softmax')
])
cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
cnn_model.fit(X_train, y_train, epochs=30, batch_size=32, verbose=2)
cnn_train_preds = np.argmax(cnn_model.predict(X_train), axis=1)
cnn_test_preds = np.argmax(cnn_model.predict(X_test), axis=1)

# --- Construct Master Labels (0 = RF better, 1 = CNN better) ---
rf_train_correct = (rf_train_preds == y_train).astype(int)
cnn_train_correct = (cnn_train_preds == y_train).astype(int)
meta_labels_train = (cnn_train_correct > rf_train_correct).astype(int)

# --- Train Master Neural Network ---
master_model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_rf_train.shape[1],)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
master_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
master_model.fit(X_rf_train, meta_labels_train, epochs=50, batch_size=16, verbose=2)

# --- Master-based Routing ---
master_decision = (master_model.predict(X_rf_test, verbose=0) > 0.5).astype(int).flatten()
master_preds = np.where(master_decision == 0, rf_test_preds, cnn_test_preds)

# --- Error Vector from Master Output ---
true_onehot = tf.keras.utils.to_categorical(y_test, num_classes=len(le.classes_))
pred_onehot = tf.keras.utils.to_categorical(master_preds, num_classes=len(le.classes_))
error_vectors = true_onehot - pred_onehot

# --- Error Prediction Neural Network ---
error_model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_rf_test.shape[1],)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(len(le.classes_))
])
error_model.compile(optimizer='adam', loss='mse')
error_model.fit(X_rf_test, error_vectors, epochs=50, batch_size=16, verbose=2)

# --- Adjusted Prediction ---
adjusted_logits = error_model.predict(X_rf_test)
adjusted_final_preds = np.argmax(pred_onehot + adjusted_logits, axis=1)

# --- Evaluation ---
rf_acc = accuracy_score(y_test, rf_test_preds)
cnn_acc = accuracy_score(y_test, cnn_test_preds)
master_acc = accuracy_score(y_test, master_preds)
final_acc = accuracy_score(y_test, adjusted_final_preds)

# --- Bar Chart with Summary ---
plt.figure(figsize=(10, 6))
plt.bar(['RF', 'CNN', 'Master', 'Error NN'], [rf_acc, cnn_acc, master_acc, final_acc],
        color=['blue', 'green', 'purple', 'orange'])
plt.ylim(0, 1)
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.grid(axis='y', linestyle='--', alpha=0.6)
summary_text = (
    f"RF Accuracy: {rf_acc:.4f} | "
    f"CNN Accuracy: {cnn_acc:.4f} | "
    f"Master Ensemble Accuracy: {master_acc:.4f} | "
    f"Error NN Adjusted Accuracy: {final_acc:.4f}"
)
plt.figtext(0.5, -0.05, summary_text, wrap=True, ha='center', fontsize=10)
plt.tight_layout()
plt.show()

# --- Utility Functions ---
def plot_actual_vs_pred(y_true, y_pred, title):
    plt.figure(figsize=(12, 6))
    cmap = plt.cm.get_cmap('tab20', len(le.classes_))
    plt.scatter(range(len(y_true)), y_true, color='black', label='Actual', marker='o', s=20, alpha=0.6)
    plt.scatter(range(len(y_pred)), y_pred, c=[cmap(p) for p in y_pred], label='Predicted', marker='x', s=20, alpha=0.7)
    plt.title(title)
    plt.ylabel("Chord Label Index")
    plt.xlabel("Sample Index")
    plt.yticks(np.arange(len(le.classes_)), le.classes_)
    r2 = r2_score(y_true, y_pred)
    plt.text(0.98, 0.02, f"R² = {r2:.2f}", ha='right', va='bottom', transform=plt.gca().transAxes,
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round'))
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred, labels=np.arange(len(le.classes_)))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le.classes_)
    disp.plot(xticks_rotation='vertical', cmap='viridis', colorbar=False)
    plt.title(title)
    plt.tight_layout()
    plt.show()

# --- Plots for All Models ---
models = {
    "Random Forest": rf_test_preds,
    "CNN": cnn_test_preds,
    "Master Ensemble": master_preds,
    "Error NN": adjusted_final_preds
}

for name, preds in models.items():
    plot_confusion_matrix(y_test, preds, f"Confusion Matrix - {name}")
    plot_actual_vs_pred(y_test, preds, f"Predicted vs Actual Labels - {name}")
```
